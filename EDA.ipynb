```python

unique_product_names = df['Product ID'].unique().tolist()  # Assuming 'Product Name' is the column name
len(unique_product_names)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Unit Price'], bins=10, kde=True)
plt.title('Distribution of Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Total Price
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Price'], bins=10, kde=True)
plt.title('Distribution of Total Price')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count of products by Brand
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Brand')
plt.title('Count of Products by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
# Quantity by Brand
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Brand', y='Quantity', estimator=sum)
plt.title('Quantity by Brand')
plt.xlabel('Brand')
plt.ylabel('Quantity')
plt.show()
```

```python
# Impact of COVID on sales
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='COVID Impact', y='Total Price')
plt.title('Impact of COVID on Total Price')
plt.xlabel('COVID Impact')
plt.ylabel('Total Price')
plt.show()
```

```python
# Seasonal impact on Quantity
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Seasonality', y='Quantity')
plt.title('Seasonal Impact on Quantity')
plt.xlabel('Seasonality')
plt.ylabel('Quantity')
plt.show()
```

```python
# Group by 'Product ID' and 'Product Name', and aggregate the 'Quantity' column
quantity_sum_df = df.groupby(['Product ID', 'Product Name'])['Quantity'].sum().reset_index()

# Save the result to a new Excel file
quantity_sum_df.to_excel("sum_of_quantities_per_product.xlsx", index=False)
```
